```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
section = ['Worst', 'Happy Medium', 'Best']
accuracy = [0.74514, 0.79541, 0.79518]
ax.bar(section, accuracy)
# plt.xlabel('Algorithms')
plt.ylim(0.7, 0.8)
plt.ylabel('Accuracy')
plt.rcParams.update({'font.size': 22})
plt.rc('figure', titlesize=22)
plt.title('RFC Comparisons')
plt.show()
```

```python
x = ['NN', 'DTC', 'RFC', 'LogR']
y1 = [.7355, .7402, .7935, .7926]
y2 = [.8014-y1[0], .7748-y1[1], .7954-y1[2], .7947-y1[3]]

plt.bar(x, y1)
plt.bar(x, y2, bottom=y1)
plt.legend(['Default', 'Optimized'])
plt.title('Optimization Comparisons')
plt.ylim(0.7, 0.85)
plt.show()
```
